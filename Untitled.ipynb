Splitting the Features and Target(cardio)

```python
X = heart_data.drop(columns='cardio', axis=1)
Y = heart_data['cardio']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Data into Training data & Test data

```python
X_train, X_test , Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training and Evaluation

Logistic Regression

```python
# Training the Logistic Regression model
model = LogisticRegression()
model.fit(X_train, Y_train)

# Predictions and accuracy evaluation for Logistic Regression
X_train_prediction = model.predict(X_train)
Y_train_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy on Training data (Logistic Regression): ', Y_train_accuracy)

X_test_prediction = model.predict(X_test)
Y_test_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy on Test data (Logistic Regression): ', Y_test_accuracy)

# Confusion Matrix for Logistic Regression
confusion_logistic = confusion_matrix(Y_test, X_test_prediction)
sns.heatmap(confusion_logistic, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
# Training the Random Forest model
rf_model = RandomForestClassifier()
rf_model.fit(X_train, Y_train)

# Predictions and accuracy evaluation for Random Forest Classifier
X_train_prediction_rf = rf_model.predict(X_train)
Y_train_accuracy_rf = accuracy_score(Y_train, X_train_prediction_rf)
print('Accuracy on Training data (Random Forest): ', Y_train_accuracy_rf)

X_test_prediction_rf = rf_model.predict(X_test)
Y_test_accuracy_rf = accuracy_score(Y_test, X_test_prediction_rf)
print('Accuracy on Test data (Random Forest): ', Y_test_accuracy_rf)

# Confusion Matrix for Random Forest Classifier
confusion_rf = confusion_matrix(Y_test, X_test_prediction_rf)
sns.heatmap(confusion_rf, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Random Forest')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Training the Decision Tree model
model_tree = DecisionTreeClassifier(random_state=42)
model_tree.fit(X_train, Y_train)

# Predictions and accuracy evaluation for Decision Tree
X_train_prediction_tree = model_tree.predict(X_train)
Y_train_accuracy_tree = accuracy_score(Y_train, X_train_prediction_tree)
print('Accuracy on Training data (Decision Tree): ', Y_train_accuracy_tree)

X_test_prediction_tree = model_tree.predict(X_test)
Y_test_accuracy_tree = accuracy_score(Y_test, X_test_prediction_tree)
print('Accuracy on Test data (Decision Tree): ', Y_test_accuracy_tree)

# Confusion Matrix for Decision Tree
confusion_tree = confusion_matrix(Y_test, X_test_prediction_tree)
sns.heatmap(confusion_tree, annot=True, fmt='d', cmap='Greens')
plt.title('Confusion Matrix - Decision Tree')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier

# Training the KNN model
model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(X_train, Y_train)

# Predictions and accuracy evaluation for KNN
X_train_prediction_knn = model_knn.predict(X_train)
Y_train_accuracy_knn = accuracy_score(Y_train, X_train_prediction_knn)
print('Accuracy on Training data (KNN): ', Y_train_accuracy_knn)

X_test_prediction_knn = model_knn.predict(X_test)
Y_test_accuracy_knn = accuracy_score(Y_test, X_test_prediction_knn)
print('Accuracy on Test data (KNN): ', Y_test_accuracy_knn)

# Confusion Matrix for KNN
confusion_knn = confusion_matrix(Y_test, X_test_prediction_knn)
sns.heatmap(confusion_knn, annot=True, fmt='d', cmap='Oranges')
plt.title('Confusion Matrix - KNN')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

# Training the Naive Bayes model
model_nb = GaussianNB()
model_nb.fit(X_train, Y_train)

# Predictions and accuracy evaluation for Naive Bayes
X_train_prediction_nb = model_nb.predict(X_train)
Y_train_accuracy_nb = accuracy_score(Y_train, X_train_prediction_nb)
print('Accuracy on Training data (Naive Bayes): ', Y_train_accuracy_nb)

X_test_prediction_nb = model_nb.predict(X_test)
Y_test_accuracy_nb = accuracy_score(Y_test, X_test_prediction_nb)
print('Accuracy on Test data (Naive Bayes): ', Y_test_accuracy_nb)

# Confusion Matrix for Naive Bayes
confusion_nb = confusion_matrix(Y_test, X_test_prediction_nb)
sns.heatmap(confusion_nb, annot=True, fmt='d', cmap='Purples')
plt.title('Confusion Matrix - Naive Bayes')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
